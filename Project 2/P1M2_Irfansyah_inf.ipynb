# Model Inference Milestone 2

## Import Libraries

```python
import numpy as np
import pickle
import random
import pandas as pd
```

## Load File

```python
# Load All Files
with open('all_process.pkl', 'rb') as file_1:
  all_process = pickle.load(file_1)
```

## Create DataFrame

```python
# Pengaturan Seed
seed_value = 69
random.seed(seed_value)
np.random.seed(seed_value)

# Generate Data
Curricular_units_2nd_sem_grade = np.random.randint(0, 19, 50)
Curricular_units_1st_sem_grade = np.random.randint(0, 19, 50)
Curricular_units_2nd_sem_approved = np.random.randint(0, 50, 50)
Curricular_units_1st_sem_approved = np.random.randint(0, 26, 50)
Curricular_units_1st_sem_evaluations = np.random.randint(0, 45, 50)
Curricular_units_2nd_sem_evaluations = np.random.randint(0, 33, 50)
Age_at_enrollment = np.random.randint(17, 70, 50)
Application_mode = np.random.randint(1, 18, 50)
Course = np.random.randint(1, 17, 50)

# Membuat DataFrame
data_inf = pd.DataFrame({
    'Curricular units 2nd sem (grade)': Curricular_units_2nd_sem_grade,
    'Curricular units 1st sem (grade)': Curricular_units_1st_sem_grade,
    'Curricular units 2nd sem (approved)': Curricular_units_2nd_sem_approved,
    'Curricular units 1st sem (approved)': Curricular_units_1st_sem_approved,
    'Curricular units 1st sem (evaluations)': Curricular_units_1st_sem_evaluations,
    'Curricular units 2nd sem (evaluations)': Curricular_units_2nd_sem_evaluations,
    'Age at enrollment': Age_at_enrollment,
    'Application mode': Application_mode,
    'Course': Course
})

data_inf
```

## Model

```python
y_pred_inf = all_process.predict(data_inf)
y_pred_inf
```

```python
# Convert into DataFrame
y_pred_inf_df = pd.DataFrame(y_pred_inf, columns=['Target - Prediction'])
y_pred_inf_df
```

```python
# Merge Inference-Set and Prediction
tabel_inference = pd.concat([data_inf, y_pred_inf_df], axis=1)
tabel_inference
```

```python
# Mengganti nilai 'Dropout' dengan 0 dan 'Graduate' dengan 1 pada kolom 'Target'
tabel_inference['Target - Prediction'] = tabel_inference['Target - Prediction'].replace({0 : 'Dropout', 1 : 'Graduate'})
```
